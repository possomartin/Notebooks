```python
def f(x): return (x**3) / 4
a, b =0,2
xp = np.linspace(a, b, 100)
yp = f(xp)

fig, ax = plt.subplots(figsize=(6,4))
ax.plot(xp, yp)

n = 10
x = np.linspace(a, b, n+1)

xcp = (x[0:-1] + x[1:]) / 2
ycp = f(xcp)
width = (b - a)/ n
width

for i in range(n):
  ax.add_patch(Rectangle((xcp[i] - width/2, 0), width, ycp[i], fill = False))


Area = width * np.sum(ycp)

Area, np.trapz(f(x), x)
```

```python

#Monte Carlo Integration: Sampling Method
fig, ax = plt.subplots(figsize=(6,4))
ax.plot(xp, yp)

n = 100 #number of rectangles
x = np.linspace(a, b, n+1) # random mid points

xcp = np.random.random(size = n) * (b - a) + a #randomly selected mid points of rectangles
ycp = f(xcp) #height at mid points of rectangles
width = (b - a)/ n
width

for i in range(n):
  ax.add_patch(Rectangle((xcp[i] - width/2, 0), width, ycp[i], fill = False))


Area = width * np.sum(ycp)
Area, np.trapz(f(x), x)
```

```python
# Monte Carlo Integration: hit and miss method

n = 100 #number of darts

fig, ax = plt.subplots(figsize=(6,4))
ax.plot(xp, yp)

c = f(b)

xr = np.random.random(size = n) * (b - a) + a
yr = np.random.random(size = n) * c


i_fuera, i_dentro = np.where(yr > f(xr)), np.where(yr < f(xr))
ax.plot(xr[i_fuera], yr[i_fuera], 'xr')
ax.plot(xr[i_dentro], yr[i_dentro], '+g')
A_rect = (b - a) * c
A_int = A_rect * len(i_dentro[0]) / n
A_int
```
