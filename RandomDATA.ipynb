```python
import requests
import json
```

```python
def GenerateRandoms(n = 1):
  for i in range(n):
    r = requests.get('https://random-data-api.com/api/users/random_user')
    data = json.loads(r.content)
    print([data["first_name"], data["last_name"], data["phone_number"], data["email"], data["address"]["street_address"], data["address"]["city"], data["address"]["state"], data["address"]["zip_code"]])
```

```python
GenerateRandoms(1000)
```
